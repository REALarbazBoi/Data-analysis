```python
#### Pandas-DataFrame And Series
'''Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. 
It provides two primary data structures: Series and DataFrame. 
A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).'''
```

```python
## Series
##A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.
```

```python
import pandas as pd
data = [x for x in range(1,6)]
series = pd.Series(data)
print("Series:\n",series)
print(type(series))
```

```python
#Create a series from Dict element
data = {"a":1,"b":2,"c":3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [x for x in range(10,31,10)]
index = ["a","b","c"]
series = pd.Series(data,index= index)
series
```

```python
# Filling missing values 
df_filled = df.fillna(0)
```

```python
df_filled.isnull().any()
```

```python
# Filling missing values with mean of the column
df["Sales_fillNA"] = df["Sales"].fillna(df["Sales"].mean())
df
```

```python
newdf = df
newdf.drop("Sales", axis = 1)
```

```python
# Back with our orignal Data
df.dtypes
```

```python
df
```

```python
# Renaming columns
df = df.rename(columns = {"Date": "Sales Date" })
df
```

```python
# Changing data types
df["Value_New"] = df["Value"].fillna(df["Value"].mean()).astype(int)
df
```

```python
df["New Value"] = df["Value"].apply(lambda x:x*2)
df.head()
```

```python
#Data Aggregating and grouping
grouped_mean = df.groupby("Product")["Value"].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(["Product", "Region"])["Value"].sum()
grouped_sum
```

```python
round(df.groupby(["Product", "Region"])["Value"].mean(),2)
```

```python
# Aggregate multiple functions
grouped_agg = df.groupby("Region")["Value"].agg(["mean", "sum", "count"])
grouped_agg
```

```python
# Merging and joining DataFrames
df1 = pd.DataFrame({"Key":["A","B","C"], "Value1": [1,2,3]})
df2 = pd.DataFrame({"Key":["A","B","D"], "Value2": [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge Data frame basis of keys
pd.merge(df1,df2,on = "Key", how = "inner")
```

```python
pd.merge(df1,df2,on = "Key", how = "outer")
```
